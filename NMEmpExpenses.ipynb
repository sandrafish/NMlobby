```python
from bs4 import BeautifulSoup
import urllib2
import csv

#the second number must be one above the last file number
for i in xrange(22, 1749):
    try:
        page = urllib2.urlopen('file:{}.html'.format(i))
    except:
        continue
    else:
        soup = BeautifulSoup(page.read())
        
#get the table with the data
        table = soup.find('table', 'reportTbl')
        #get the rows
        list_of_rows = []
        for row in table.findAll('tr')[1:]:
            col = row.findAll('td')
            emp = col[0].text.strip()
            filing = col[1].text.strip()
            status = col[2].text.strip()
            submitted = col[3].text.strip()
            cont = col[4].text.strip()
            exp = col[5].text.strip()
            record = (i, emp, filing, status, cont, exp)
            print emp
            #write to file
            writer = csv.writer(open('emp.csv', 'ab'))
            writer.writerow(record)
```
