Step 3: switch to mainlob folder (above) and run script to get a csv file with the data, below. Make sure to change the second number to one above the last number in your set of files. That number also gives you an idea of how much to step up the numbers in the next wget.

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import urllib2
import csv

#the second number must be one above the last file number
for i in xrange(21, 1657):
    try:
        page = urllib2.urlopen('file:{}.html'.format(i))
    except:
        continue
    else:
        soup = BeautifulSoup(page.read())
        #get the name
        n = soup.find(id='ctl00_ContentPlaceHolder1_lnkBCLobbyist')
        name = n.string
        print name
#get the table with the data
        table = soup.find('table', 'reportTbl')
        #get the rows
        for row in table.findAll('tr')[1:]:
            col = row.findAll('td')
            filing = col[0].string
            status = col[1].string
            filed = col[2].text.strip()
            cont = col[3].text.strip()
            exp = col[4].text.strip()
            record = (i, name, filing, status, filed, cont, exp)
            #write to file
            writer = csv.writer(open('lobby.csv', 'ab'))
            writer.writerow(record)
    
```

```python
emplob17 = lobemp17.groupby(['lob_name', 'lob_id'])['employer'].apply(lambda x: "{%s}" % ', '.join(x)).reset_index()
```
