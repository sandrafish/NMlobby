```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
from bs4 import BeautifulSoup
import urllib2
import csv

#the second number must be one above the last file number
for i in xrange(22, 6610):
    try:
        page = urllib2.urlopen('file:{}.html'.format(i))
    except:
        continue
    else:
        soup = BeautifulSoup(page.read())
        #get the name
        n = soup.find(id='ctl00_ContentPlaceHolder1_lblName')
        name = n.string
        print name
#get the table with the data
        table = soup.find('table', 'reportTbl')
        #get the rows
        for row in table.findAll('tr')[2:]:
            col = row.findAll('td')
            eventdate = col[0].string
            eventtype = col[1].string
            recipient = col[2].string
            total = col[3].text.strip()
            record = (i, name, eventdate, eventtype, recipient, total)
            #write to file
            writer = csv.writer(open('events.csv', 'ab'))
            writer.writerow(record)
```
